Add to the dataframe a new feature which will describe whether or not the user have done more then 3 service calls. Investigate how does this feature describes the churn rate. Create a simple (baseline) rule for identifying loyal users by using only the features 'Inernational plan' and newly constructed feature. Calculate the accuracy (percentage) of the prediction by means of this rule. How many users which are loyal are marked as churn? How many users which are churn are marked as loyal?

```python
df['>3 service calls'] = (df['Customer service calls']>3).astype(int)
sns.countplot(x='>3 service calls', hue='Churn', data=df);
pd.crosstab(df['>3 service calls'], df['Churn'], margins=True)

# simple baseline

df['baseline']=(df['>3 service calls']&df['International plan']).astype(int)
pd.crosstab(df['baseline'], df['Churn'], margins=True)

# accuracy
acc = float(np.sum(df['baseline']==df['Churn']))/df.shape[0]
acc
```

```python
from sklearn import decomposition
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score

pca = decomposition.PCA(n_components=15)
# now we are using the 

X = df[features_full].as_matrix()
#print df[features_full]

y = df['Churn'].as_matrix()
# centralisizing the data
X_centered = X - X.mean(axis=0)
X_centered /=X.std(axis=0)

pca.fit(X_centered)
X_pca = pca.transform(X_centered)

for i, component in enumerate(pca.components_):
    print("{} component: {}% of initial variance".format(i + 1, 
          round(100 * pca.explained_variance_ratio_[i], 2)))
    print(" + ".join("%.3f x %s" % (value, name)
                     for value, name in zip(component,
                                            df[features_full].columns)))
```

```python
pca = decomposition.PCA().fit(X_pca)

plt.figure(figsize=(10,7))
plt.plot(np.cumsum(pca.explained_variance_ratio_), color='k', lw=2)
plt.xlabel('Number of components')
plt.ylabel('Total explained variance')
plt.xlim(0, 15)
plt.yticks(np.arange(0, 1.1, 0.1))
```
